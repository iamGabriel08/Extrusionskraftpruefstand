## Plot von Extrusionskraft Messung

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datei im selben Ordner wie das Notebook:
file_name = "Force Measurement.csv"

# CSV einlesen:
# - engine="python" erlaubt das automatische Finden von ; oder ,
# - comment=None damit nichts wegkommentiert wird
# - wir überspringen die erste Zeile, falls sie "Parameters:" enthält
with open(file_name, "r", encoding="utf-8", errors="replace") as f:
    first_line = f.readline()

skiprows = 1 if first_line.strip().lower().startswith("parameters:") else 0

df = pd.read_csv(
    file_name,
    sep=None,           # erkennt ; oder , automatisch
    engine="python",
    skiprows=skiprows
)

# Spaltennamen vereinheitlichen
df.columns = [c.strip().lower() for c in df.columns]

# In Zahlen umwandeln (falls als Text eingelesen)
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna()

# --- Einzelplots ---
plt.figure(figsize=(10,4))
plt.plot(df["time"], df["temperature"])
plt.xlabel("Time [s]")
plt.ylabel("Temperature [°C]")
plt.title("Temperature vs Time")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(df["time"], df["force"])
plt.xlabel("Time [s]")
plt.ylabel("Force [N]")
plt.title("Force vs Time")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,4))
plt.plot(df["time"], df["slip"])
plt.xlabel("Time [s]")
plt.ylabel("Slip [%]")
plt.title("Slip vs Time")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
